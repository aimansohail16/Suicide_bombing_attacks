# Missing Values Calculation

```python
data.isnull().sum()
```

# Handle Missing Values

```python
cat_col1 = ['Holiday Type', 'Targeted Sect if any']
for col1 in cat_col1:
    data[col1] = data[col1].fillna('None')
cat_col2 = ['Explosive Weight (max)', 'Influencing Event/Event', 'Time', 'Hospital Names', 'Open/Closed Space', 'Blast Day Type','Location']
for col2 in cat_col2:
    data[col2] = data[col2].fillna('Unknown')
num_col1 = ['Killed Min', 'Injured Min', 'Injured Max']
for col3 in num_col1:
    data[col3] = data[col3].fillna(0)
num_col2 = ['Temperature(C)', 'No. of Suicide Blasts']
for col4 in num_col2:
    data[col4] = data[col4].fillna(data[col4].median())
loc_data = ['Latitude', 'Longitude']
for col5 in loc_data:
    data[col5] = data[col5].fillna(data.groupby(['Province', 'City'])[col5].transform('mean'))
    data[col5] = data[col5].fillna(data[col5].mean())
```

```python
data.isnull().sum()
```

# Statistical Summary

```python
data.describe()
```

```python
data["Date"]=data["Date"].str.replace("-"," ")
data.head()
```

```python
#Lets seperate day month and year from date
day_name=data.Date.str.split(' ').str[0]
month=data.Date.str.split(' ').str[1]
day=data.Date.str.split(' ').str[2]
year=data.Date.str.split(" ").str[3]
data["Day_name"]=day_name
data["Month"]=month
data["Day"]=day
data["Year"]=year
data.drop("Date",axis=1,inplace=True)
data.head()
```
